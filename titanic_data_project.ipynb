# EDA is Starts

## what is the over all survival rate

```python
over_all_survival_rate=train["Survived"].value_counts().reset_index(name="Count")
over_all_survival_rate
```

```python
sns.barplot(data=over_all_survival_rate,x="Survived",y="Count")
plt.xticks([0,1],labels=["Not Survived","Survived"])
plt.show()
```

## what is the survival rate by passanger class

```python
survival_rate_by_pclass=train.groupby("Pclass")["Survived"].value_counts().unstack()
survival_rate_by_pclass
```

```python
survival_rate_by_pclass.plot(kind="bar",stacked=True)
plt.show()
```

## what is the survival by gender

```python
survival_by_gender=train.groupby("Sex")["Survived"].size().reset_index(name="Death")
survival_by_gender
```

## what is the survival by Embarked point

```python
survival_by_embarked=train.groupby("Embarked")["Survived"].mean().reset_index()
survival_by_embarked
```

```python
sns.barplot(data=survival_by_embarked,x="Embarked",y="Survived")
plt.show()
```

## what is the Gender Distribution of Passengers

```python
gender_dist_passengers=train["Sex"].value_counts().reset_index(name="count")
gender_dist_passengers
```

```python
plt.pie(gender_dist_passengers["count"],labels=["Male","Female"],autopct="%0.3f%%")
plt.title("Gender Distribution of Passengers")
plt.ylabel('')  # Remove the default ylabel for better visualization
plt.legend(labels=["Male","Female"], title="Gender", loc="best")
plt.tight_layout()
plt.show()
```

## what is the survival rate by gender and age group

```python
survival_rate_by_gender_age=train.groupby(["Sex","Age"])["Survived"].value_counts()
survival_rate_by_gender_age
```

```python
age_bins = [0, 12, 18, 35, 60, 100]  # Age groups: 0-12, 13-18, 19-35, 36-60, 61+
age_labels = ['Child', 'Teen', 'Young Adult', 'Adult', 'Senior']

# Add a new column for age groups to the DataFrame
train['AgeGroup'] = pd.cut(train['Age'], bins=age_bins, labels=age_labels, right=False)
```

```python
survival_rate_gender_age = train.groupby(['Sex', 'AgeGroup'])['Survived'].mean().unstack()
survival_rate_gender_age
```

```python
plt.figure(figsize=(12, 8))
sns.violinplot(
    data=train, 
    x='AgeGroup', 
    y='Survived', 
    hue='Sex', 
    split=True, 
    inner='quartile', 
    palette={'female': '#ff9999', 'male': '#66b3ff'}
)

# Customizing the plot
plt.title("Survival Rate Distribution by Gender and Age Group")
plt.xlabel("Age Group")
plt.ylabel("Survival Rate")
plt.legend(title="Gender", loc="upper right")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

# Age based Analysis

## what is the distribution of passenger age

```python
mean_age = train['Age'].mean()
median_age = train['Age'].median()
mode_age = train['Age'].mode()[0]
age_range = (train['Age'].min(), train['Age'].max())

print(f"Mean Age: {mean_age:.2f}")
print(f"Median Age: {median_age:.2f}")
print(f"Mode Age: {mode_age}")
print(f"Age Range: {age_range}")
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(
    data=train,  
    x="Age", 
    bins=30,  
    kde=True,
    color="skyblue",
    edgecolor="black",
    alpha=0.7
)


plt.title("Passenger Age Distribution", fontsize=16)
plt.xlabel("Age", fontsize=12)
plt.ylabel("Frequency", fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

# Correlations and Relationship

## what is the correlations between numarical features?

```python
numerical_features = ["Survived",'Age', 'Fare', 'Pclass',]
correlation_of_numatical=train[numerical_features].corr()
correlation_of_numatical
```

```python
sns.heatmap(correlation_of_numatical, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix for Titanic Numerical Features')
plt.show()
```
